# 實作股價K線圖
台積電 TSMC (2330) <br>
聯發科 MediaTek (2454) 

```python
import yfinance as yf
import datetime as dt

# 設置日期範圍
start = dt.datetime(2024, 1, 1)
end = dt.datetime(2024, 10, 8)

# 取得台積電 (2330.TW) 的股價資料
df_2330 = yf.download('2330.TW', start=start, end=end)


# 取得聯發科 MediaTek (2454.TW) 的股價資料
df_2454 = yf.download('2454.TW', start=start, end=end)

# 顯示前幾行數據
print(df_2330.tail())
print(df_2454.tail())
```

### 台積電 VS 聯發科 股價線圖

```python
import matplotlib.pyplot as plt
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()

fig = plt.figure(figsize=(10, 6))
plt.plot(df_2330.Close, label='TSMC')
plt.plot(df_2454.Close, label='MediaTek')
plt.title('TSMC vs MediaTek', loc='center')
plt.xlabel('Date')
plt.ylabel('Price')
plt.grid(True, axis='y')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(10, 6))
plt.plot(df_2330.Close, label='TSMC')
plt.plot(df_2454.Close, label='MediaTek')
plt.title('TSMC V.S. MediaTek', loc = 'center')
plt.xlabel('Date')
plt.ylabel('Price')
plt.grid(True, axis='y')
plt.legend()
plt.show()
```
